#### Question 4
#### Interpreting Sensitivity, Specificity and Model accuracy:
#### Sensitivity:
> Sensitivity (also called the true positive rate, or the recall in some fields) measures the proportion of actual positives which are correctly identified as such, and is complementary to the false negative rate. Sensitivity= true positives/(true positive + false negative).
#### Specificity:
> Specificity (also called the true negative rate) measures the proportion of negatives which are correctly identified as such, and is complementary to the false positive rate. Specificity=true negatives/(true negative + false positives).
#### Accuracy:
> Model accuracy is defined as the number of classifications a model correctly predicts divided by the total number of predictions made.

#### How should we use cook's distance and standardise residual in logistic regression?

>Cook’s distance, D<sub>i</sub>, is used in Regression Analysis to find influential outliers in a set of predictor variables. In other words, it’s a way to identify points that negatively affect your regression model. The measurement is a combination of each observation’s leverage and residual values; the higher the leverage and residuals, the higher the Cook’s distance.
<br>
>Several interpretations for Cook’s distance exist. There isn’t a universally accepted rule for cut
off points.<br>
>One interpretation is to investigate any point over 4/n, where n is the number of
observations.<br>
>Other authors suggest that any “large” Di should be investigated. How large is “too
large”? The consensus seems to be that a Di value of more than 1 indicates an influential
value, but you may want to look at values above 0.5. Any value that sticks out from the
other (like the one in the above chart) should also be investigated.<br>
>An alternative (but slightly more technical) way to interpret Di is to find the potential
outlier’s percentile value using the F-distribution. A percentile of over 50 indicates a
highly influential point.<br>
>If you have a lot of points with large Di values, that could indicate a problem with your
regression model in general.<br>

# QUESTION 4 — Sensitivity, Specificity and Accuracy

```python
from sklearn.metrics import confusion_matrix

#Finding the threshold were the loss is minimum using the Maximum Likehood Estimation(MLE)
thresholds = np.arange(0,1,0.01)

loss_list = []

for th in thresholds:
    y_th = np.where(y_train_pred_prob > th, 1, 0)
    tn, fp, fn, tp = confusion_matrix(y_train, y_th).ravel()

    # Log loss / Negative Log Likelihood
    # adding small value to avoid log(0)
    loss = -(tp*np.log(th+1e-10) + tn*np.log(1-th+1e-10))
    loss_list.append(loss)

#Finding the threshold with minimum loss
min_loss = min(loss_list)
best_threshold = thresholds[loss_list.index(min_loss)]

print("Best Threshold using MLE:", best_threshold)
```
